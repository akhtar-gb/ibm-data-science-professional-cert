### Importing Libraries and Setting Up Data:

```python
# python
# Copy code
# This script generates a random dataset and fits a linear regression model to it.
# It then plots the data and the fitted line.

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Random data generation
rng = np.random.RandomState(42) # Seed for reproducibility
x = 10 * rng.rand(50) # 50 random x values in range 0 to 10
y = 2 * x - 1 + rng.randn(50) # y values with some noise

# Scatter plot of the data
plt.scatter(x, y, color='blue', label='Data')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Randomly Generated Data")
plt.legend()
plt.show()
```

### Create and Fit the Model:

```python
# Reshape x into a features matrix (50 samples, 1 feature)
X = x[:, np.newaxis]

# Create a linear regression model
model = LinearRegression()
model.fit(X, y) # Fit the model to the data

# Retrieve slope (coef_) and intercept (intercept_)
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope: {slope:.3f}, Intercept: {intercept:.3f}")
```

### Make Predictions:

```python
# Generate new unseen data
x_fit = np.linspace(0, 10, 100) # 100 evenly spaced points between 0 and 10
X_fit = x_fit[:, np.newaxis] # this line reshapes x_fit into a 2D array - the array contains 100 samples and 1 feature
# print x_fit array to see what it looks like
print("x_fit array:", x_fit)

# Predict y values for new data
y_fit = model.predict(X_fit)

# Visualize original data and fitted line
plt.scatter(x, y, color='blue', label='Original Data')
plt.plot(x_fit, y_fit, color='red', label='Fitted Line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression Fit")
plt.legend()
plt.show()
```
